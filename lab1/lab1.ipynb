## Activity Two

```python
import altair as alt
```

```python
scatter_plot = alt.Chart(cars).mark_circle().encode(
    x = 'displ',
    y = 'hwy'
)

scatter_plot.show()
```

```python
scatter_plot_colored = scatter_plot.encode(
    color = 'class'
)

scatter_plot_colored.show()
```

```python
scatter_plot = alt.Chart(cars).mark_circle().encode(
    x = 'displ',
    y = 'hwy',
    color = 'hwy',
    size = 'hwy'
)

scatter_plot.show()
```
