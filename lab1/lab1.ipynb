## Activity Two

```python
scatter_plot = alt.Chart(cars).mark_circle().encode(
    x = 'displ',
    y = 'hwy'
)

scatter_plot.show()
```

```python
scatter_plot_colored = scatter_plot.encode(
    color = 'class'
)

scatter_plot_colored.show()
```

```python
# cars_audi = cars[cars['manufacturer'] == 'audi']

scatter_plot = alt.Chart(cars).mark_circle().encode(
    x = 'displ',
    y = 'hwy',
    color = 'hwy',
    size = 'hwy'
)

scatter_plot.show()
```

## Activity Three

```python
cars['mpg'] = (cars['cty'] + cars['hwy']) / 2

mean_mpg_by_manufacturer = cars.groupby('manufacturer')['mpg'].mean().reset_index()

bar_chart = alt.Chart(mean_mpg_by_manufacturer).mark_bar().encode(
    x = alt.X('manufacturer:N', sort='-y', title='manufacturer'),
    y = alt.Y('mpg:Q', title='Mean MPG')
).properties(
    width=600,
    height=400,
    title='Mean MPG by Manufacturer'
)

bar_chart.show()
```

## Activity Four

```python
histogram = alt.Chart(cars).mark_bar().encode(
    x = alt.X('displ:Q'),
    y = alt.Y('count()', title='Number of Cars')
).properties(
    width=600,
    height=400,
    title='Distribution of Engine Displacement'
)

histogram.show()
```

```python
box_plot = alt.Chart(cars).mark_boxplot().encode(
    x = alt.X('class:N', title='Car Class'),
    y = alt.Y('hwy:Q', title='Highway MPG'),
    color = alt.Color('class:N', legend=None),
    size=alt.SizeValue(50)
).properties(
    width=600,
    height=400,
    title='Box Plot of Highway MPG by Car Class'
)

box_plot.show()
```
